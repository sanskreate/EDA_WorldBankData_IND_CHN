# üìä Dataset Description: Socio-Economic and Development Indicators (1999‚Äì2023)

This dataset provides annual country-level statistics from **1999 to 2023** for two of Asia‚Äôs largest economies ‚Äî **India** and **China**. #DATA CLEANING AND PREPROCESSING

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.stats import zscore
```

```python
df = pd.read_csv('/content/WorldDevelopmentData_INDCHN.csv')
df
```

```python
df.columns = ['country_name', 'country_code', 'series_name'] + [f'year_{y}' for y in range(1999, 2024)]
df.head()
```

```python
year_cols = [col for col in df.columns if col.startswith('year_')]
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
df['country_name'] = df['country_name'].str.strip()
df['country_code'] = df['country_code'].str.strip()
df['series_name'] = df['series_name'].str.strip()
```

```python
# reshaping the dataset from wide to long format for year-wise analysis
df_melted = df.melt(
    id_vars=['country_name', 'country_code', 'series_name'],
    var_name='year',
    value_name='value'
)
df_melted['year'] = df_melted['year'].str.extract(r'(\d+)').astype(int)
df_melted
```

```python
# pivoting the dataset to have indicators as columns for each country and year
df_pivot = df_melted.pivot_table(
    index=['country_name', 'year'],
    columns='series_name',
    values='value'
).reset_index()
df_pivot
```

```python
df_pivot.info()
```

```python
# filling missing values with forward-fill and backward-fill grouped by country
df_pivot.update(df_pivot.groupby('country_name').ffill())
df_pivot.update(df_pivot.groupby('country_name').bfill())
df_pivot
```

```python
df_pivot.isna().sum()
```

# OUTLIER DETECTION AND ANALYSIS

```python
for country in ['India', 'China']:
    country_df = df_pivot[df_pivot['country_name'] == country]
    z_scores = zscore(country_df.select_dtypes('float'))
    anomalies = (abs(z_scores) > 3).sum(axis=0)
    print(f'\n{country} - Anomalies count by feature:')
    print(pd.Series(anomalies, index=country_df.select_dtypes('float').columns).sort_values(ascending=False))
```

```python
# analysing hospital beds per 1,000 people over time (India) (anamoly detected)

plt.figure(figsize=(10, 5))
sns.lineplot(data=df_pivot[df_pivot['country_name'] == 'India'],
             x='year', y='Hospital beds (per 1,000 people)')
plt.title('Hospital Beds per 1,000 People - India')
plt.grid(True)
plt.show()
```

###Based on the line graph showing Hospital Beds per 1,000 People in India from approximately 1998 to 2022, barious findings are:

1. Despite fluctuations, India has experienced a general downward trend in hospital bed density over the two-decade period, declining from approximately 2.1 beds per 1,000 people in 1998 to about 1.6 beds per 1,000 people by 2022.

2. The graph shows two dramatic temporary drops in hospital bed availability:
   - Around 2002-2004: dropping from about 2.0 to 0.9 beds per 1,000 people
   - Around 2012-2013: an even more severe drop to approximately 0.5 beds per 1,000 people

3. The overall reduction from about 2.1 to 1.6 beds per 1,000 people represents approximately a 24% decline in per capita hospital capacity, potentially indicating significant healthcare access challenges as India's population has grown.

```python
# percentage change analysis of indicators like inflation, mortality rates, hospital beds, and internet access over the years: India vs China

sns.set_style("whitegrid")
features = [
    'Inflation, consumer prices (annual %)',
    'Mortality rate, under-5 (per 1,000 live births)',
    'Hospital beds (per 1,000 people)',
    'Individuals using the Internet (% of population)'
]
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, feature in zip(axes.flatten(), features):
    df_pivot[f'{feature}_change'] = df_pivot.groupby('country_name')[feature].pct_change() * 100

    sns.histplot(
        data=df_pivot,
        x=f'{feature}_change',
        hue='country_name',
        bins=30,
        kde=True,
        ax=ax,
        multiple='layer'
    )
    ax.set_title(f'% Change: {feature}', fontsize=10)

plt.tight_layout()
plt.show()
```

###Based on the histogram plots comparing percentage changes in key metrics between China and India, various findings are:

1. Inflation patterns (top-left):
   - India shows more extreme inflation changes, with several instances of high positive inflation (peaking around 8-9%).
   - China demonstrates more moderate inflation changes, mostly clustered between -100% to +100%.

2. Under-5 mortality rate changes (top-right):
   - China's improvements are concentrated in the -6 to -8 range (larger reductions).
   - India's improvements are more widely distributed between -4 to -6, suggesting less consistent progress.

3. Hospital beds per 1,000 people (bottom-left):
   - China shows more positive changes concentrated around +25, indicating consistent increases in hospital capacity
   - India has instances of very large positive changes (+250), suggesting periods of rapid expansion from a lower base

4. Internet usage growth (bottom-right):
   - Both countries show similar patterns of internet adoption
   - The highest frequency of changes for both countries is in the 0-20% range
